```python
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
from openpmd_viewer.addons import LpaDiagnostics
```

```python
ts = LpaDiagnostics('./diags/diag/')
```

# Visualization

```python
def visualize_iteration(iteration):

    plt.clf()

    plt.subplot(121)
    rho, info = ts.get_field('rho_electrons', iteration=iteration)
    plt.imshow(abs(rho.T), cmap='Blues',
               extent=[1e6*info.zmin, 1e6*info.zmax, 1e6*info.rmin, 1e6*info.rmax],
               aspect='auto')
    Ex, info = ts.get_field('E', 'x', iteration=iteration)
    plt.imshow(abs(Ex.T), cmap='gist_heat_r',
               extent=[1e6*info.zmin, 1e6*info.zmax, 1e6*info.rmin, 1e6*info.rmax],
               aspect='auto', alpha=0.5)
    plt.xlabel('$z [μm]$')
    plt.ylabel('$x [μm]$')


    plt.subplot(122)
    #spect, info_spect = ts.get_spectrogram(iteration=iteration, pol='x')
    #plt.imshow(spect[:,::-1], cmap='gist_heat_r', extent=[info.zmin, info.zmax, info_spect.omegamin, info_spect.omegamax], aspect='auto')
    #plt.ylim(0, 1e16)

    ts.get_spectrum(iteration=iteration, pol='x', plot=True, color='r')
    plt.xlim(0, 0.5e16)
    plt.ylim(0,0.015)
    plt.grid()

    plt.savefig('iteration_%d.png' % iteration)

    plt.tight_layout()
```

```python
ts.iterate( visualize_iteration )
```
