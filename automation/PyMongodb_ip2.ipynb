```python
## This notebook includes simulation and experimental data
## in a database using PyMongo
## Author : Revathi Jambunathan
## Date : January, 2025


import pymongo
import pandas as pd
```

```python
# Connection to the Mongo Client
db = pymongo.MongoClient(host="", 
                             username="",
                             password="", 
                             authSource="")[""]
```

```python
# print names of collections
for collection_name in db.list_collection_names():
    print(collection_name)
```

```python
def process_csv(csv_file, expt_flag, date):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # Add new column to denote if its expt (1) or simulation (0)
    df['experiment_flag'] = expt_flag
    
    # Add new column to include date
    df['date'] = date

    return df
```

```python
# Define path to experimental and simulation data
file_path='/global/homes/r/rjnathan/Codes/2024_IFE-superfacility'
sim_csv=file_path+'/simulation_data/simulation_data.csv'
expt_csv=file_path+'/experimental_data/experimental_data.csv'

# get ip2 collection
ip2_collection=db["ip2"]

# read simulation data and include date and flag
df_sim = process_csv(sim_csv, 0, "01-12-2025")
# load the simulatation data
data_dict = df_sim.to_dict(orient='records')
ip2_collection.insert_many(data_dict)

# read simulation data and include date and flag
df_expt = process_csv(expt_csv, 1, "03-03-2024")
# load the experimental data
exptdata_dict = df_expt.to_dict(orient='records')
ip2_collection.insert_many(exptdata_dict)
```
